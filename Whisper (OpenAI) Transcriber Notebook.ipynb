```python
from openai import OpenAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv() # Load environment variables from .env file
```

```python
api_key = os.getenv("OPENAI_API_KEY")
```

```python
client = OpenAI(api_key=api_key)
```

```python
with open(r"C:\Users\PC\My_drive\data_scientist\Whisper_01\audio_small.mp3", "rb") as f:
    resp = client.audio.transcriptions.create(
        model="whisper-1",
        file=f,
        language="ru"      # можно убрать, но так надёжнее
    )

text = resp.text
print(text)
```

```python
with open("transcript_whisper.txt", "w", encoding="utf-8") as out:
    out.write(text)
```
